## Contexto: Superposição Quântica

No Inatel, dentro da disciplina de F04 (Física Moderna e Quântica) há um conceito fundamental: a Superposição Quântica. Este é um dos pilares da Computação Quântica ao possibilitar a combinação de dois estados simultâneos. O cálculo da função de onda é essencial para determinar o estado de um sistema quântico. Uma análise da relação entre a posição de estudo $x$ pode ser modelada pela função:

$$\psi(x)= \frac{1}{\sqrt{2}} (\sin{x} + \cos{x})$$

$$\psi'(x)= \frac{1}{\sqrt{2}}(\cos{x} - \sin{x})$$

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x): return (1/np.sqrt(2))*(np.sin(x) + np.cos(x))
def df(x): return (1/np.sqrt(2))*(np.cos(x) - np.sin(x))
```

```python
def newton_raphson(x0,tol=1e-9,max_iter=100):
    x=x0
    for i in range(max_iter):
        if abs(f(x))<tol: return x,i
        x=x-f(x)/df(x)
    return x,max_iter
```

## Parte Prática
1. Plote f(x) dentro do intervalo [1;5], indicando a linha de 0 tracejada interceptando a curva.

```python
x = np.linspace(1,5,400)
y = f(x)

plt.plot(x,y)
plt.axhline(0, linestyle='--')
plt.grid()
plt.show()
```

2. Aplique Newton-Raphson com x0 = 2

```python
raiz, it = newton_raphson(2)
print("Raiz:", raiz)
print("Iterações:", it)
```

3. Teste diferentes chutes iniciais (0.2, 0.4, 1.5, 1.8, 3.5, 5.0) e compare o número de iterações.

```python
chutes = [0.2,0.4,1.5,1.8,3.5,5.0]

for x0 in chutes:
    raiz, it = newton_raphson(x0)
    print(x0, "->", raiz, "| iter:", it)
```

4. Modifique a tolerância e analise o impacto na precisão.

```python
tols = [1e-3,1e-6,1e-9,1e-12]

for tol in tols:
    raiz, it = newton_raphson(2, tol=tol)
    print("tol =", tol, "| raiz =", raiz, "| iter =", it)
```

1. O que significa convergência quadrática?

**Resposta:**
Convergência quadrática significa que o erro diminui muito rapidamente, sendo proporcional ao quadrado do erro anterior. Na prática, a precisão dobra a cada iteração próxima da raiz.

2. Como adaptar o método sem derivada?

**Resposta:**
Pode-se usar o método das secantes, que aproxima a derivada usando dois pontos, eliminando a necessidade de calcular f'(x).

3. Compare este método com o da bissecção, indicando as vantagens no uso de cada um.

**Resposta:**
Newton-Raphson é mais rápido (convergência quadrática), porém depende do chute inicial. Bissecção é mais lento (linear), mas sempre converge se houver mudança de sinal.
